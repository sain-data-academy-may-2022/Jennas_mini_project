```python
import numpy as np
import matplotlib.pyplot as plt 

food_name = %sql SELECT food_name FROM food
quantity = %sql SELECT quantity FROM food
quantity1 = int(quantity[0][0])
quantity2 = int(quantity[1][0])
quantity3 = int(quantity[2][0])
quantity4 = int(quantity[3][0])
quantity5 = int(quantity[4][0])
quantity_list = [quantity1, quantity2, quantity3, quantity4, quantity5]
print(quantity_list)
```

```python
plt.figure(figsize=(18, 7))
x_data = food_name
y_data = quantity_list


x_pos = np.arange(len(x_data))
plt.bar(x_pos, y_data)
plt.xticks(x_pos, x_data)
plt.ylabel('Stock')
plt.title('Stock Chart')
plt.show()
```
